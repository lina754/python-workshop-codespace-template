# Pydantic Settings: very simple demo

This shows the basic mechanics of `BaseSettings`:
- define fields with defaults
- load values from `demo_config.toml`
- override with environment variables (`APP_` prefix)

```python
import os
from pathlib import Path

from pydantic_settings import BaseSettings, SettingsConfigDict


class AppSettings(BaseSettings):
    model_config = SettingsConfigDict(env_prefix="APP_")

    app_name: str = "claims-default"
    timeout_seconds: int = 30
    debug: bool = False


config_path = Path("demos/demo_config.toml")

settings = AppSettings(_toml_file=config_path)
print("Loaded from TOML:")
print(settings.model_dump())

os.environ["APP_DEBUG"] = "true"
settings_with_env = AppSettings(_toml_file=config_path)
print("\nWith env override (APP_DEBUG=true):")
print(settings_with_env.model_dump())
os.environ.pop("APP_DEBUG", None)
```
