# Tiny Standard Library Tour

A very short, case-agnostic side note for your workshop.

These are small examples of standard-library tools that are useful in many data science scripts.

```python
from pathlib import Path
from datetime import date
import shutil

print("Today:", date.today())

base = Path("tmp_path_demo")

base.mkdir(exist_ok=True)
file_path = base / "note.txt"

file_path.write_text("hello from pathlib\n", encoding="utf-8")
print("read_text():", file_path.read_text(encoding="utf-8").strip())

with file_path.open("a", encoding="utf-8") as f:
    f.write("second line\n")

print("glob('*.txt'):", [p.name for p in base.glob("*.txt")])

#cleanup
shutil.rmtree(base)
```

## `collections` + `itertools`

Quick counting and compact transformations.

```python
from collections import Counter
from itertools import islice, chain

values = ["A", "B", "A", "C", "B", "A"]
counts = Counter(values)
print("Counts:", counts)
print("Top 2:", counts.most_common(2))

# chain combines iterables without nesting.
combined = list(chain([1, 2], [3], [4, 5]))
print("Chained:", combined)

# First 3 pairs from a tiny generator expression.
pairs = ((i, i**2) for i in range(10))
print("Sample pairs:", list(islice(pairs, 3)))
```

## `statistics` + `math`

Lightweight numeric summaries before using heavier libraries.

```python
from statistics import mean, median
from math import log1p

x = [1, 2, 2, 3, 5, 8]
print("Mean:", mean(x))
print("Median:", median(x))
print("log1p transform:", [round(log1p(v), 3) for v in x])
```
